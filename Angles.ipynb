The measure of these angles must always be $90^\text{o}$, therefore we call them *complimentary*. Screenshot 2022-04-06 10.21.33 AM.png

There are two ways to show why that must always be true. Taking a look at them will help us review our knowledge of angle facts of both parallel lines as well as triangles.

Recall that the measures of all the interior angles of a triangle must sum to $180^\text{o}$. Since horizontal is always perpindicular vertical by definition, our triangle will always be a right triangle with the right angle located where the altitude line intersects the ground (ideally horozontal). Therefore the two acute angles of a right triangle must always be complimentary

We can also observe that angles $\angle{D}\text{ and }\angle{C}$ are complimentary as well. We know this must be true from our previous observation that $\angle{E}\cong\angle{D}$, so then by the Transitive Property if $\angle{E}\text{ and }\angle{D}$ are congruent, and $\angle{E}\text{ and }\angle{C}$ are complimentary then angles $\angle{D}\text{ and }\angle{C}$ must be complimentary as well.
